# Matplotlib

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

# Plotting a single point
plt.figure(0)
plt.plot(10,200,'r*')
plt.show()
```

```python
#Line plot
x=np.linspace(2,9,500)
y=np.linspace(2,9,500)*4
plt.plot(x,y,'r')
plt.show()
```

```python
# Ploting curve
x=np.linspace(-10,10,10000)
y=x
plt.grid('on')
plt.title('circle')
plt.plot(x,np.sqrt(100-y**2),'r')
plt.plot(x,-np.sqrt(100-y**2),'r')
plt.xlim(-11,11)
plt.ylim(-11,11)
plt.show()                 
```

```python
# Scatter plot
x=np.random.randn(500)
y=np.random.randn(500)
plt.scatter(x,y,c='r')
plt.show()
```

```python
# Plotting sin and cosine wave
x=np.linspace(-2*np.pi,2*np.pi,1000)
y1=np.sin(x)
y2=np.cos(x)
plt.xlim(-2*np.pi,2*np.pi)
plt.plot(x,y1,'r',label='y=sin(x)')
plt.plot(x,y2,'g',label='y=cos(x)')
plt.legend(loc=10)
plt.show()
```

```python
# Subplot 

#Line plot
#plt.figure(1)
plt.subplot(2,2,1)
x=np.linspace(2,9,500)
y=np.linspace(2,9,500)*4
plt.title('Line')
plt.plot(x,y,'r')

# Ploting circle
#plt.figure(2)
plt.subplot(2,2,2)
x=np.linspace(-10,10,10000)
y=x
plt.title('circle')
plt.plot(x,np.sqrt(100-y**2),'r')
plt.plot(x,-np.sqrt(100-y**2),'r')
plt.xlim(-11,11)
plt.ylim(-11,11)


# Scatter plot
#plt.figure(3)
plt.title('Scatter plot')
plt.subplot(2,2,3)
x=np.random.randn(500)
y=np.random.randn(500)
plt.scatter(x,y,c='r')


# Plotting sin and cosine wave
#plt.figure(4)
plt.title('Sin and cosine')
plt.subplot(2,2,4)
x=np.linspace(-2*np.pi,2*np.pi,1000)
y1=np.sin(x)
y2=np.cos(x)
plt.xlim(-2*np.pi,2*np.pi)
plt.plot(x,y1,'r',label='y=sin(x)')
plt.plot(x,y2,'g',label='y=cos(x)')
plt.legend(loc=10)
plt.show()
```
